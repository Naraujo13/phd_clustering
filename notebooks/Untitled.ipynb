```python
import pandas as pd
```

```python
programs = pd.read_json('../data/program_keywords.json')
```

```python
programs.head()
```

```python
programs = programs.loc[programs.keywords.apply(lambda c: c!=[])]
```

```python
len(programs)
```

```python
terms = set()
for index, row in programs.iterrows():
    terms.update(row['keywords'].keys())
```

```python
len(terms)
```

```python
p_copy = programs
```

```python
for term in terms:
    p_copy[term] = 0
```

```python
for index, row in programs.iterrows():
    for term in programs.loc[index, 'keywords'].keys():
        programs.loc[index, term] = programs.loc[index, 'keywords'][term]
```

```python
programs.head()
```

```python
programs['id'] = programs.index
```

```python
programs.head()
```

```python
programs.to_csv('../data/program_keywords.csv')
```

```python
programs_numeric = programs.drop(columns=['institutions', 'keywords', 'name'])
```
